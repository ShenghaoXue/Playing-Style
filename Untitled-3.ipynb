# Data Dictionary

## 1. Identity

- **Player** – Player name
- **Comp** - League

---

## 2. Shooting

- **Goals** – Goals scored
- **Shots** – Total shots
- **SoT** – Shots on target
- **ShoDist** – Average shot distance
- **ShoFK** – Shots from free kicks
- **ShoPK** – Penalty goals
- **PKatt** – Penalty kicks attempted

---

## 3. Passing

- **PasTotDist** – Total passing distance
- **PasTotPrgDist** – Progressive passing distance
- **PasShoCmp/PassShoAtt** – Short passes completed/attempted
- **PasMedCmp/PassMedAtt** – Medium passes completed/attemped
- **PasLonCmp/PassLonAtt** – Long passes completed/attemped

---

## 4. Attack Progression

- **Assists** – Assists
- **PasAss** – Key passes
- **Pas3rd** – Passes into attacking third
- **PPA** – Passes into penalty area
- **CrsPA** – Crosses into penalty area
- **PasProg** – Progressive passes
- **TB** – Through balls
- **Sw** – Switches
- **PasCrs** – Crosses
- **TI** – Throw-in taken
- **CK** – Corner kicks taken

---

## 5. Shot-Creating Actions (SCA): actions directly leading to a shot

- **ScaPassLive** – From live passes
- **ScaPassDead** – From set pieces
- **ScaDrib** – From dribble
- **ScaSh** – From shot rebound
- **ScaFld** – From foul drawn
- **ScaDef** – From defensive action

---

## 6. Goal-Creating Actions (GCA): actions directly leading to a goal

- **GcaPassLive / Dead / Drib / Sh / Fld / Def** – same as SCA

---

## 7. Defense

- **Tkl** – Tackles
- **TklDef3rd** - Tackles won in defensive third
- **TklMid3rd** - Tackles won in mid third
- **Att3rd** – Tackles won in attacking third
- **TklDri** – Tackles vs dribbler
- **TklDriAtt** – Dribble attempts faced
- **TklDriPast** – Dribbled past
- **Blocks** – Total blocks
- **BlkSh** – Shot blocks
- **BlkPass** – Pass blocks
- **Int** – Interceptions
- **Clr** – Clearances
- **Err** – Errors leading to shot/goal
- **Recov** – Ball recoveries
- **AerWon** – Aerial duels won
- **AerLost** – Aerial duels lost
- **AerWon%** – Aerial success rate

---

## 8. Possession

- **Touches** – Total touches
- **TouDefPen** - Total touches in own penality area
- **Def3rd** - Total touches in defensive third
- **Mid3rd** - Total touches in mid third
- **Att3rd** - Total touches in attacking third
- **TouAttPen** - Total touches in opponenent penalty area
- **TouLive** – Live-ball touches
- **ToAtt** – Dribble attempts
- **ToSuc** – Successful dribbles
- **ToTkl** – Times tackled while dribbling
- **Carries** – Total carries
- **CarTotDist** – Total carry distance
- **CarPrgDist** – Progressive carry distance
- **CarProg** – Progressive carries
- **Car3rd** – Carries into final third
- **CPA** – Carries into penalty area
- **CarMis** – Miscontrols while carrying
- **CarDis** – Dispossessed while carrying
- **Rec** – Passes received
- **RecProg** – Progressive passes received

---

## 9. Discipline & Fouls

- **CrdY** – Yellow cards
- **CrdR** – Red cards
- **2CrdY** – Second yellow leading to red
- **Fls** – Fouls committed
- **Fld** – Fouls drawn
- **Off** – Offsides
- **OG** – Own goals
